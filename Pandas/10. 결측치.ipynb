## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원 번호')
df
```

```python
df.dropna(inplace=True) # 전체 데이터 중에서 nan을 포함하는 데이터 삭제 
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원 번호')
df
```

- axis : index or columns
- how : any or all

```python
df.dropna(axis='index', how='any') # nan 하나라도 있는 행 삭제
```

```python
df.dropna(axis='columns', how='any') # nan이 하나라도 있는 열 삭제
```

```python
df.dropna(axis='index', how='all')
```

```python
df.dropna(axis='columns', how='all')
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all') # 열 데이터 전체가 nan인 경우에만 열 삭제
```
