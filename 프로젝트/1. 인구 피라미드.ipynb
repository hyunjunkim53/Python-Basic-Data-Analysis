# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
file_name = '202512_202512_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y') # skiprows를 통해 처음 3개의 행 생략, 행정기관을 index로 지정, usecols를 통해 필요한 특정 행만 뽑아옴
df_m.head(3) # 위 3개의 행만 출력
```

```python
print(df_m.index.values)
```

```python
df_m.loc['전국  ']
```

```python
df_m.iloc[0]
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,155,577 -> 1155577 (정수형), 숫자에 쉼표가 들어가면 문자열로 인식하기 때문에
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns # Pandas에서는 열 이름 중복 불가능 따라서 열 이름 중복 방지 위해 자동으로 .1 를 붙여 구분
```

```python
df_w.columns = df_m.columns # 컬럼명 동일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 윈도우
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명, 남자 데이터는 0을 기준으로 왼쪽으로
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 단위 : 천 명, 여자 데이터는 0을 기준으로 오른쪽으로
plt.title('2025 대한민국 인구 피라미드')
plt.savefig('2025_인구피라미드.png', dpi=100)
plt.show()
```
